### **1. k-평균을 사용할 때 적절한 클러스터 개수를 선택할 수 있는 두 가지 기법을 설명하세요.**
___



엘보우 메서드 (Elbow Method): 엘보우 메서드는 클러스터 개수(k)에 대한 오차 제곱합(Sum of Squared Errors, SSE)을 계산하고 그래프로 나타내는 방법입니다. SSE는 각 데이터 포인트와 해당 클러스터 중심과의 거리를 제곱하여 모든 클러스터에 대해 합한 값입니다. 엘보우 메서드에서는 k를 증가시키면서 SSE가 어떻게 변하는지 관찰합니다. 그래프를 그린 후, 클러스터 개수가 증가함에 따라 SSE가 빠르게 감소하다가 둔화되는 지점을 찾습니다. 이 지점이 "팔꿈치(Elbow)" 모양과 유사하게 나타나는 경우, 해당 지점의 k 값을 적절한 클러스터 개수로 선택합니다. 이 지점에서 SSE가 감소하기 시작하는데, 이는 클러스터의 중심과 데이터 포인트 사이의 거리를 더 최소화할 필요가 없어졌음을 의미합니다.

실루엣 스코어 (Silhouette Score): 실루엣 스코어는 클러스터 내 데이터 포인트와 클러스터 간 데이터 포인트 사이의 거리를 기반으로 클러스터 품질을 측정하는 지표입니다. 실루엣 스코어는 각 데이터 포인트에 대해 실루엣 계수를 계산하고 이를 모든 데이터 포인트에 대해 평균을 구합니다. 실루엣 계수는 -1에서 1 사이의 값을 가지며, 클러스터 품질을 나타냅니다. 값이 1에 가까울수록 클러스터 내 데이터 포인트가 서로 밀접하게 모여 있고, 클러스터 간 데이터 포인트는 떨어져 있는 것을 나타냅니다. 일반적으로 실루엣 스코어가 높을수록 클러스터가 더 좋은 품질을 가집니다. 클러스터 개수(k)를 조정하면서 실루엣 스코어를 계산하고, 스코어가 최대가 되는 k 값을 선택합니다.

### **2. 대규모 데이터셋으로 확장할 수 있는 군집 알고리즘 두 개를 말해보세요. 밀도가 높은 지역을 찾는 군집 알고리즘 두 개는 무엇인가요?**
___


MiniBatch K-Means: K-Means 알고리즘은 대규모 데이터셋에서도 잘 동작하지만 계산 비용이 높을 수 있습니다. MiniBatch K-Means는 이러한 문제를 완화하기 위한 변형 알고리즘으로, 데이터셋을 작은 미니배치로 나누어 클러스터링을 수행합니다. 이로써 전체 데이터셋을 사용하는 것보다 효율적인 계산이 가능하며, 대규모 데이터셋에서 빠른 수행이 가능합니다.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise): 밀도가 높은 지역을 찾는데 특히 유용한 군집 알고리즘 중 하나입니다. DBSCAN은 데이터 포인트 주변에 밀도가 높은 지역을 찾아 클러스터를 형성합니다. 이 알고리즘은 데이터 포인트 간 거리를 기반으로 클러스터를 구성하며, 잡음 데이터를 구분할 수 있는 능력이 있습니다. 대규모 데이터셋에서도 잘 동작하며, 클러스터의 크기나 모양에 대한 가정을 하지 않기 때문에 다양한 형태의 클러스터를 찾을 수 있습니다.

### **3. 가우시안 혼합이 무엇인가요? 어떤 작업에 사용할 수 있나요?**
___


가우시안 혼합(Gaussian Mixture)은 확률 분포를 모델링하는 데 사용되는 통계적 모델 중 하나입니다. 이 모델은 여러 개의 가우시안(정규 분포) 확률 분포의 선형 결합으로 데이터를 모델링합니다. 각 가우시안 확률 분포는 데이터의 특정 구성 요소 또는 패턴을 나타내며, 이러한 확률 분포들이 섞여서 데이터셋을 설명합니다. 가우시안 혼합 모델은 데이터가 어떤 혼합된 확률 분포로부터 생성되었다고 가정하며, 이 가정을 통해 데이터의 구조와 클러스터를 추론할 수 있습니다.

가우시안 혼합 모델은 주로 다음과 같은 작업에 사용됩니다:

클러스터링(Clustering): 가우시안 혼합 모델은 데이터 포인트를 여러 개의 클러스터로 그룹화할 때 사용됩니다. 각 클러스터는 가우시안 확률 분포로 모델링되며, 데이터 포인트는 각 클러스터에 할당됩니다. 이를 통해 데이터의 클러스터 구조를 발견할 수 있습니다.

이상치 탐지(Anomaly Detection): 가우시안 혼합 모델은 데이터셋의 정상적인 패턴을 학습하고, 이상치나 이상 동작을 탐지할 때 사용됩니다. 모델은 대부분의 데이터가 일반적인 패턴을 따르고, 드물게 발생하는 이상치를 식별할 수 있습니다.

확률 밀도 추정(Probability Density Estimation): 가우시안 혼합 모델은 데이터셋의 확률 밀도 함수를 추정하는 데 사용됩니다. 이를 통해 새로운 데이터 포인트가 어떤 확률 분포에서 생성되었을 가능성을 계산할 수 있습니다.

차원 축소(Dimensionality Reduction): 가우시안 혼합 모델을 사용하여 데이터의 저차원 표현을 추정할 수 있습니다. 이를 통해 데이터의 중요한 특성을 추출하거나 시각화에 활용할 수 있습니다.

가우시안 혼합 모델은 확률적 모델링과 클러스터링 문제에 유용하며, 데이터의 분포를 통계적으로 모델링하고 다양한 응용 분야에서 활용됩니다.



### **4. 다음 단계를 따라 올리베티 얼굴 데이터셋을 군집해보자..**
___
전통적인 올리베티 얼굴 데이터셋은 64 X 64 픽셀 크기의 흑백 얼굴 이미지 400개를 가지고 있습니다. 사람 40명의 사진을 10장씩 찍은 것인데, 각 이미지는 4096 크기의 1D 벡터로 펼쳐져 있습니다.

#### **STEP 1. 아래 코드를 실행하여 데이터셋을 불러오세요.**

```python
from sklearn.datasets import fetch_olivetti_faces

olivetti = fetch_olivetti_faces()
```

####  **STEP 2. 데이터셋을 훈련 세트, 검증 세트, 테스트 세트로 나누세요.**
해당 데이터셋은 매우 작으므로, `StratifiedShuffleSplit` 함수를 이용하여, 계층적 샘플링으로 각 세트에 동일한 사람의 얼굴이 고루 섞이도록 하세요.

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

strat_split=StratifiedShuffleSplit(n_splits=1,test_size=40,random_state=42)
train_valid_idx,test_idx=next(strat_split.split(olivetti.data,olivetti.target))
X_train_valid=olivetti.data[train_valid_idx]
y_train_valid=olivetti.target[train_valid_idx]
X_test=olivetti.data[test_idx]
y_test=olivetti.target[test_idx]

strat_split=StratifiedShuffleSplit(n_splits=1,test_size=80,random_state=43)
train_idx,valid_idx=next(strat_split.split(X_train_valid,y_train_valid))
X_train=X_train_valid[train_idx]
y_train=y_train_valid[train_idx]
X_valid=X_train_valid[valid_idx]
y_valid=y_train_valid[valid_idx]

print(X_train.shape,y_train.shape)
print(X_valid.shape,y_valid.shape)
print(X_test.shape,y_test.shape)
```

####  **STEP 3. 속도를 높이기 위해 각 학습/검증/테스트 피처 데이터셋에 PCA을 적용하여 데이터의 차원을 줄이세요.**
- `n_componets = 0.99`로 설정합니다.

```python
from sklearn.decomposition import PCA

pca=PCA(0.99)
X_train_pca=pca.fit_transform(X_train)
X_valid_pca=pca.transform(X_valid)
X_test_pca=pca.transform(X_test)

pca.n_components_
```

####  **STEP 4. 교재에서 다룬 방법을 사용하여 적절한 클러스터 개수를 찾고 k-평균을 사용해 이미지를 군집해보세요.**

```python
from sklearn.cluster import KMeans

k_range = range(5, 150, 5)
kmeans_per_k = []
for k in k_range:
    print("k={}".format(k))
    kmeans = KMeans(n_clusters=k, random_state=42).fit(X_train_pca)
    kmeans_per_k.append(kmeans)
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores=[silhouette_score(X_train_pca,model.labels_)
for model in kmeans_per_k]
best_index=np.argmax(silhouette_scores)
best_k=k_range[best_index]
best_score=silhouette_scores[best_index]

import matplotlib.pyplot as plt

plt.figure(figsize=(8,3))
plt.plot(k_range,silhouette_scores,'bo-')
plt.xlabel('k')
plt.ylabel('Silhouette score')
plt.plot(best_k,best_score,'rs')
plt.show()
```

```python
best_k
```

```python
inertias=[model.inertia_ for model in kmeans_per_k]
best_inertia=inertias[best_index]

plt.figure(figsize=(8,3.5))
plt.plot(k_range,inertias,'bo-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.plot(best_k,best_inertia,'rs')
plt.show()
```

####  **STEP 5. 아래 `plot_faces` 함수를 이용하여, 클러스터를 시각화하고, 각 클러스터에 비슷한 얼굴이 들어있는지 확인해보세요.**

```python
# plot_faces 함수
def plot_faces(faces, labels, n_cols=5):
    faces = faces.reshape(-1, 64, 64)
    n_rows = (len(faces) - 1) // n_cols + 1
    plt.figure(figsize=(n_cols, n_rows * 1.1))
    for index, (face, label) in enumerate(zip(faces, labels)):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(face, cmap="gray")
        plt.axis("off")
        plt.title(label)
    plt.show()
```

```python
best_model=kmeans_per_k[best_index]
for cluster_id in np.unique(best_model.labels_):
  print('cluster',cluster_id)
  in_cluster=best_model.labels_ == cluster_id
  faces=X_train[in_cluster]
  labels=y_train[in_cluster]
  plot_faces(faces,labels)
```
